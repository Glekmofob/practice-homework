```python
def get_gcd(num1: int, num2: int) -> int:
    while max(num1, num2) % min(num1, num2) != 0:
        if num1 > num2:
            num1 = num1 % num2
        else:
            num2 = num2 % num1
    return min(num1, num2)


print(get_gcd(1071, 147))
```

```python
def IsPrime(num: int) -> bool:
    d = 2
    while num % d != 0:
        d += 1
    return d == num


def get_sum_of_prime_divisors(num: int) -> int:
    sum_of_divisors = 0
    divisors = []
    i = 2
    while num != 1:
        if num % i == 0:
            if IsPrime(i):
                divisors.append(i)
                num //= i
                i = 2

        else:
            i += 1
    divisors = set(divisors)
    for i in divisors:
        sum_of_divisors += i
    return sum_of_divisors


print(get_sum_of_prime_divisors(12))
```

```python
def is_palindrome(num: int) -> bool:
    if num < 0:
        return False
    num_reversed = 0
    num_reversed += num % 10
    num_current = num
    num_current //= 10
    while num_current > 0:
        num_reversed *= 10
        num_reversed += num_current % 10
        num_current //= 10
    return num == num_reversed


print(is_palindrome(121))
```

```python
def flip_bits_in_range(num: int, left_bit: int, right_bit: int) -> int:
    answer = 0
    position = 1
    curr_bit = 1
    while num > 0:
        if curr_bit >= left_bit and curr_bit <= right_bit:
            if num % 2 == 1:
                answer = answer + position * 0
            else:
                answer = answer + position * 1
        else:
            answer = answer + position * (num % 2)
        position *= 10
        num //= 2
        curr_bit += 1
    num = 0
    power = 0
    print(answer)
    while answer > 0:
        digit = answer % 10
        num += digit * (2**power)
        answer //= 10
        power += 1
    return num


print(flip_bits_in_range(0b1011, 2, 2))
```

```python
def get_nth_digit(num: int) -> int:
    count_digit = 1
    flag_finish = False
    answer = 0
    i = 0
    while True:
        k = i * 2
        while k > 0:
            if num == count_digit:
                answer = k % 10
                flag_finish = True
                break
            k //= 10
            count_digit += 1
        if flag_finish:
            break
        i += 1
    return answer


get_nth_digit(5)
```

```python
def merge_intervals(intervals: list[list[int, int]]) -> list[list[int, int]]:
    intervals.sort(key=lambda x: x[0])
    if len(intervals) == 0:
        return []
    newintervals = []
    start, finish = intervals[0][0], intervals[0][1]
    flag1 = True
    for i in intervals:
        if flag1:
            flag1 = False
            continue  # скип первого элемента(костыль)
        if start < i[0] and finish > i[1]:
            continue
        elif finish >= i[0] and start <= i[0]:
            finish = i[1]
        else:
            newintervals.append([start, finish])
            start = i[0]
            finish = i[1]
    newintervals.append([start, finish])
    return newintervals


print(merge_intervals([[1, 4], [2, 3], [1, 4]]))
```

```python
def unzip(compress_text: str) -> str:
    result = ""
    compress_text = compress_text.split(" ")
    for i in compress_text:
        if "*" not in i:
            result += i
        else:
            result += i[:-2] * int(i[-1])
    return result


print(unzip("a*10"))
```

```python
def reg_validator(reg_expr: str, text: str) -> bool:
    i, j = 0, 0
    while i < len(reg_expr) and j < len(text):
        symbol = reg_expr[i]
        if symbol == "d":
            if not text[j].isdigit():
                print(i)
                return False
            while j < len(text) and text[j].isdigit():
                j += 1
        elif symbol == "p":
            if not text[j].isalpha():
                return False
            while j < len(text) and text[j].isalpha():
                j += 1
        elif symbol == "s":
            if not (text[j].isalpha() or text[j].isdigit()):
                return False
            while j < len(text) and (text[j].isalpha() or text[j].isdigit()):
                j += 1
        else:
            if text[j] != symbol:
                return False
            j += 1
        i += 1

    return i == len(reg_expr) and j == len(text)


print(reg_validator("w", "hello"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

L = 10.0
g = 9.81
Omega = 7.292115e-5
phi = np.deg2rad(55)

omega0 = np.sqrt(g / L)
omega_p = Omega * np.sin(phi)

dt = 0.05
t = np.arange(0, 500, dt)

A = 1.0
x = A * np.cos(omega0 * t) * np.cos(omega_p * t)
y = A * np.cos(omega0 * t) * np.sin(omega_p * t)

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')
ax.set_xlim(-1.2*A, 1.2*A)
ax.set_ylim(-1.2*A, 1.2*A)

line, = ax.plot([], [], lw=2)
point, = ax.plot([], [], 'ro')

trail_x, trail_y = [], []

def update(frame):
    trail_x.append(x[frame])
    trail_y.append(y[frame])
    line.set_data(trail_x, trail_y)
    point.set_data([x[frame]], [y[frame]])
    return line, point

ani = FuncAnimation(fig, update, frames=len(t), interval=30)
plt.show()
```

```python
from dataclasses import dataclass
from math import acos, isclose


@dataclass(eq=False, init = False)
class Vector2D:
    _abscissa: float
    _ordinate: float 

    def __init__(self, abscissa: float = 0., ordinate: float= 0.):
        self._abscissa = abscissa
        self._ordinate = ordinate
    
    
    def __repr__(self):
        return (f"Vector2D(abscissa={self._abscissa}, ordinate={self._ordinate})")

    @property
    def abscissa(self):
        return self._abscissa

    @property
    def ordinate(self):
        return self._ordinate

    def __ne__(self, other: "Vector2D"):
        return not (self == other)

    def __eq__(self, other):
        if not isinstance(other, Vector2D):
            return NotImplemented
        return isclose(self.abscissa, other.abscissa, rel_tol=1e-9, abs_tol=1e-10) and isclose(
            self.ordinate, other.ordinate, rel_tol=1e-9, abs_tol=1e-10)

    def __lt__(self, other):
        if not isinstance(other, Vector2D):
            return NotImplemented
        return (
            self.abscissa < other.abscissa
            and not (isclose(self.abscissa, other.abscissa, rel_tol=1e-9, abs_tol=1e-10))
        ) or (
            isclose(self.abscissa, other.abscissa, rel_tol=1e-9, abs_tol=1e-10)
            and self.ordinate < other.ordinate
            and not (isclose(self.ordinate, other.ordinate)))

    def __le__(self, other):
        return self == other or self < other

    def __gt__(self, other):
        return not (self <= other)

    def __ge__(self, other):
        return not (self < other)

    def __abs__(self):
        return (self.abscissa**2 + self.ordinate**2) ** 0.5

    def __bool__(self):
        return not (isclose(abs(self), 0, abs_tol=1e-12))

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            return Vector2D(self.abscissa * other, self.ordinate * other)
        return NotImplemented

    def __rmul__(self, other):
        return self * other

    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            return Vector2D(self.abscissa / other, self.ordinate / other)
        return NotImplemented

    def __add__(self, other):
        if isinstance(other, Vector2D):
            return Vector2D(self.abscissa + other.abscissa, self.ordinate + other.ordinate)
        if isinstance(other, (int, float)):
            return Vector2D(self.abscissa + other, self.ordinate + other)
        return NotImplemented

    def __radd__(self, other):
        return self + other

    def __sub__(self, other):
        if isinstance(other, Vector2D):
            return Vector2D(self.abscissa - other.abscissa, self.ordinate - other.ordinate)
        if isinstance(other, (int, float)):
            return Vector2D(self.abscissa - other, self.ordinate - other)
        return NotImplemented

    def __neg__(self):
        return self * -1

    def __float__(self):
        return abs(self)

    def __int__(self):
        return int(abs(self))

    def __complex__(self):
        return complex(self.abscissa, self.ordinate)

    def __matmul__(self, other: "Vector2D"):
        if not isinstance(other, Vector2D):
            return NotImplemented
        return self.abscissa * other.abscissa + self.ordinate * other.ordinate

    def get_angle(self, other: "Vector2D") -> float:
        if not isinstance(other, Vector2D):
            raise TypeError("Other must be Vector2D")
        if isclose(abs(other), 0) or isclose(abs(self), 0):
            raise ValueError("You cannot get angle in between given vector and a zero vector")
        return acos((self @ other) / (abs(self) * abs(other)))

    def conj(self):
        return Vector2D(self.abscissa, self.ordinate * -1)

print(Vector2D(2,1))
```

```python
class MyEnumerate:
    def __init__(self, iterable, start=0):
        # Сохраняем итератор из переданного объекта
        self.iterator = iter(iterable)
        self.count = start

    def __iter__(self):
        # Итератор должен возвращать самого себя
        return self

    def __next__(self):
        # 1. Получаем следующее значение из вложенного итератора
        # Если элементы закончатся, iter() сам выбросит StopIteration,
        # и наш генератор тоже остановится.
        value = next(self.iterator)
        
        # 2. Формируем текущий результат
        result = (self.count, value)
        
        # 3. Увеличиваем счетчик для следующего шага
        self.count += 1
        
        return result

# Проверка:
names = ["Алексей", "Мария", "Иван"]
for idx, name in MyEnumerate(names, start=4):
    print(f"{idx}: {name}")
```

```python
def giga_any(iterable):
    iterable = iter(iterable)
    count = 0
    for item in iterable:
        count+=1
        if item:
            return item
    if count == 0:
        return False
    return item
print(giga_any([0,1]))
        
```
